# How to convert images into structured arrays and scalars?

First, let's download a sample image, for example: `https://www.pexels.com/photo/woman-in-collared-shirt-774909/`

Let's import the cv2 (for reading an image from disk) and matplotlib (for plotting the loaded image).

```
%matplotlib inline
import cv2, matplotlib.pyplot as plt
img = cv2.imread('/content/pexels-andrea-piacquadio-774909.jpg')
```

Let's take a look at this image.

```
plt.imshow(img)
```

Let's crop the image between 150th-600th rows, as well as 100th-500th columns and then convert the image into grayscale using the following
code and plot it:


```
img = img[150:600,100:500]
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

Let's Convert the image into a 50 x 50 array and plot it:

```
img_gray_small = cv2.resize(img_gray,(50,50))
plt.imshow(img_gray_small, cmap='gray')
```

 Let's have a look at the pixel values.

```
print(img_gray_small)
```

Let's reordered the channels using the cv2.cvtcolor method.

```
img = cv2.imread('/content/pexels-andrea-piacquadio-774909.jpg')
img = img[150:600,100:500,:]
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
print(img.shape)
```

Let's crop the image.

```
crop = img[100:,100:]
plt.imshow(crop)
```
